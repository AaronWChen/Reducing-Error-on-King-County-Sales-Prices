```python
impactful = ['sqft_living', 'dist_from_bellevue', 'dist_from_seattle', 'water_weight', 'view_weight', 'sqft_times_grade']
```

```python
for feature in impactful:
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(1,1,1)
    ax.scatter(x=test_data[feature], y=log_test_prices)
    ax.set_ylabel("Ln[ $ USD ]", fontsize=16)
    ax.set_xlabel(f"{feature}", fontsize=16)
    ax.set_title(f"Natural Log of Price vs {feature}", fontsize=24)
    plt.savefig(f"../write_data/ln_price_vs_{feature}.png")
    plt.show()
```
